## Importamos librerías a utilizar

```python
#Para crear y análizar redes
import networkx as nx
#Para dibujar redes y graficar funciones
import matplotlib.pyplot as plt
```

## Creación de Distintas redes

#### Creamos una red completa de 100 nodos

```python
K_100 = nx.complete_graph(100)
```

#### Creamos una red aleatoria de 100 nodos y probabilidad de conexión de 0.06 - (Algoritmo de Gilbert con parámteros n=100, p=0.06)

```python
redAleatoria = nx.erdos_renyi_graph(100, 0.06, seed=None, directed=False)
```

#### Creamos una red de mundo pequeño con 100 nodos y probabilidad de reconexión de p=0.3 (Algoritmo de Watts-Strotgatz con parámetros n=100, k=4, p=0.3

```python
redMundoPeque = nx.watts_strogatz_graph(100, 4, 0.3, seed=None)
```

#### Creamos una red de escala libre con 100 nodos y 2 enlaces por cada nuevo nodo agregado a la red (Algoritmo Barabsi-Albert n=100, 2)

```python
redEscalaLibre = nx.barabasi_albert_graph(100, 2, seed=None)
```

## Información de cada red

### Red Completa

#### Calculamos la información

```python
#number_of_nodes() devuelve la cantidad de nodos contenidos en la red completa K_100
numero_nodos_k = K_100.number_of_nodes()

#number_of_edges() devuelve la cantidad de enlaces contenidos en la red completa K_100
numero_enlaces_k = K_100.number_of_edges()

#diameter() devuelve el diámetro de la red completa K_100
diameter_k = nx.diameter(K_100)

#density() devuelve la densidad de la red completa K_100
densidad_k = nx.density(K_100)
```

#### Imprimimos la información

```python
print ("Nodos: " + str(numero_nodos_k))
print ("Enlaces: " + str(numero_enlaces_k))
print ("Diametro: " + str(diameter_k))
print ("Densidad: " + str(densidad_k))
```

```python
#number_of_nodes() devuelve la cantidad de nodos contenidos en la red aleatoria
numero_nodos_redAleatoria = redAleatoria.number_of_nodes()

#number_of_edges() devuelve la cantidad de enlaces contenidos en la red aleatoria
numero_enlaces_redAleatoria = redAleatoria.number_of_edges()

#diameter() devuelve el diámetro de la red aleatoria
diameter_redAleatoria = nx.diameter(redAleatoria)

#density() devuelve la densidad de la red aleatoria
densidad_redAleatoria = nx.density(redAleatoria)
```

#### Imprimimos la información

```python
print ("Nodos: " + str(numero_nodos_redAleatoria))
print ("Enlaces: " + str(numero_enlaces_redAleatoria))
print ("Diametro: " + str(diameter_redAleatoria))
print ("Densidad: " + str(densidad_redAleatoria))
```

## Distribución de Grados

#### Calculamos y gráficamos la distribución de grados de cada red

```python
#En la Variable dist_K se calcula la distribución de grados de una red aleatoria
dist_K = nx.degree_histogram(K_100)

#En la Variable dist_redAleatoria se calcula la distribución de grados de una red aleatoria
dist_redAleatoria = nx.degree_histogram(redAleatoria)

#En la Variable dist_redMundoPeque se calcula la distribución de grados de una red de mundo pequeño
dist_redMundoPeque = nx.degree_histogram(redMundoPeque)

#En la Variable dist_redEscalaLibre se calcula la distribución de grados de una red de escala libre
dist_redEscalaLibre = nx.degree_histogram(redEscalaLibre)
```

#### Graficamos las distribuciones

```python
# Las siquientes 4 gráficas muestran la distribución de grados de la red completa K_100, de la red aleatoria,
#de la red de mundo pequeño y de la red de escala libre 

#Se crea una figura con dos filas y dos columnas
fig, axs = plt.subplots(2, 2)

#Con fig.suptitle se le da un titulo a la figura de 2 filas y 2 columnas
fig.suptitle("Distribución de grados", y=1.05)

#En la posición [0,0] de la figura imprime la gráfica de la distribución de grados de la red completa K_100
axs[0, 0].plot(dist_K)
axs[0, 0].set_title('Red Completa')

#En la posición [0,1] de la figura imprime la gráfica de la distribución de grados de la red aleatoria
axs[0, 1].plot(dist_redAleatoria)
axs[0, 1].set_title('Red Aleatoria')

#En la posición [1,0] de la figura imprime la gráfica de la distribución de grados de la red de mundo pequeño
axs[1, 0].plot(dist_redMundoPeque)
axs[1, 0].set_title('Red de Mundo Pequeño')

#En la posición [1,1] de la figura imprime la gráfica de la distribución de grados de la red de escala libre
axs[1, 1].plot(dist_redEscalaLibre)
axs[1, 1].set_title('Red de Escala libre')

#Para cada una de las gráficas se le asigna un titulo a cada "eje x" y a cada "eje y" 
for ax in axs.flat:
    ax.set(xlabel='Grado', ylabel='# Nodos')

#tight_layout() permite que se vean todos los títulos de las graficas sin que se corten
fig.tight_layout()
#savefig permite guardar en una imagen (.png) la figura con las 4 gráficas
plt.savefig("dist_degree.png")
```

## Dibujando las Redes

```python
#draw() de la libreria networkX permite dibujar la gráfica de la red de escala libre 
nx.draw(redEscalaLibre, node_size=100, node_color="g")
#savefig permite guardar en una imagen (.png) la gráfica dibujada
plt.savefig("scaleFree.png")
```
